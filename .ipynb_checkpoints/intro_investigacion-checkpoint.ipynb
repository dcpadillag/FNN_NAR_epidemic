#**MODELO SIR: Italia antes de la cuarentena**

```python
dates = (dates - dates[0]).astype('timedelta64[D]').astype(int)
dates_pre= dates[:14]
# Tiempo
t = np.arange(len(data_italy_pre))

# Modelo SIR
def sir_model(y, t, beta, gamma):
    S, I, R = y
    dSdt = (-beta * S * I)/60000000
    dIdt = ((beta * S * I)/60000000) - gamma * I
    dRdt = gamma * I
    return np.array([dSdt, dIdt, dRdt])

def rk4(f, y0, t, beta, gamma):
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    for i in range(1, len(t)):
        h = t[i] - t[i - 1]
        k1 = f(y[i - 1], t[i - 1], beta, gamma)
        k2 = f(y[i - 1] + h * k1 / 2, t[i - 1] + h / 2, beta, gamma)
        k3 = f(y[i - 1] + h * k2 / 2, t[i - 1] + h / 2, beta, gamma)
        k4 = f(y[i - 1] + h * k3, t[i - 1] + h, beta, gamma)
        y[i] = y[i - 1] + (h / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    return y


# Función de pérdida (mínimos cuadrados)
'''
def loss(params):
    beta, gamma = params
    y0 = [1 - data_italy_pre[0], data_italy_pre[0], 0.0]
    sol = odeint(sir_model, y0, t, args=(beta, gamma))
    I_pred = sol[:, 1]
    return np.sum((I_pred - data_italy_pre)**2)
'''
def loss(params):
    beta, gamma = params
    y0 = [1 - data_italy_pre[0], data_italy_pre[0], 0.0]
    sol = rk4(sir_model, y0, t, beta, gamma)
    I_pred = sol[:, 1]
    return np.sum((I_pred - data_italy_pre) ** 2)


# Estimación de parámetros
bounds = [(0.0001, 0.6), (0.0001, 0.06)]
x0 = [0.3,0.03]
#result = differential_evolution(loss, bounds)
result = minimize(loss, x0, method='SLSQP', bounds=bounds)
beta_opt, gamma_opt = result.x

print(f"\n--- Italia ---")
print(f"Beta óptimo: {beta_opt}")
print(f"Gamma óptimo: {gamma_opt}")
print(f"R0: {beta_opt / gamma_opt}")

# Simular con los valores óptimos
y0 = [1 - data_italy_pre[0], data_italy_pre[0], 0.0]
sol = odeint(sir_model, y0, t, args=(beta_opt, gamma_opt))
I_fit = sol[:, 1]

# Plot
plt.figure(figsize=(10,6))
plt.plot(t, data_italy_pre, 'ro', label='Datos reales (Infectados)')
plt.plot(t, I_fit, 'b-', label='SIR Ajustado')
plt.xlabel("Días desde inicio")
plt.ylabel("Proporción de infectados")
plt.title("Ajuste SIR al brote en Italia")
plt.legend()
plt.grid()
plt.show()
```

#**MODELO SIR: Italia después de la cuarentena**


```python
from scipy.special import expit

def simulate_sir_optimal_control(data, dates, beta, gamma):
    """Simula el modelo SIR con control óptimo H(t) y guarda resultados cada subpaso (dt = 0.2)."""
    Nt = len(data) - 1
    dt = 0.2
    steps_per_day = int(1 / dt)
    total_steps = Nt * steps_per_day

    # Inicialización de arrays extendidos (cada subpaso)
    S = np.zeros(total_steps + 1)
    I = np.zeros(total_steps + 1)
    R = np.zeros(total_steps + 1)
    H = np.zeros(Nt)

    # Condiciones iniciales
    S_temp = 1 - data[0]
    I_temp = data[0]
    R_temp = 0

    S[0] = S_temp
    I[0] = I_temp
    R[0] = R_temp

    index = 1
    for n in range(Nt):
        # Control óptimo H(t) calculado una vez por día
        xn = (data[n] - I[n] + I[n] * gamma) / (I[n] * S[n] * beta)
        H[n] = np.clip(xn, 0, 1)
        #H[n] = expit(xn)
        #H[n] = (np.arctan(xn) / np.pi) + 0.5
        #S[n+1] = S[n] - dt*(beta*I[n]*S[n]*H[n])
        #I[n+1] = I[n] + dt*(beta*I[n]*S[n]*H[n]-gamma*I[n])
        #R[n+1] = R[n] + dt*(gamma*I[n])
        
        for _ in range(steps_per_day):
            dS = -dt * beta * I_temp * S_temp * H[n]
            dI = dt * (beta * I_temp * S_temp * H[n] - gamma * I_temp)
            dR = dt * gamma * I_temp

            S_temp += dS
            I_temp += dI
            R_temp += dR

            S[index] = S_temp
            I[index] = I_temp
            R[index] = R_temp
            index += 1
        
    return S, I, R, H


def plot_results(dates, data, I, H, country="Country"):
    """Grafica resultados comparando datos y simulación con control H(t)."""
    fig, axs = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

    # Gráfico superior: Datos vs Simulación
    axs[0].plot(dates, data, 'r.', label='Datos reportados')
    axs[0].plot(dates, I[::5], 'b-', label='Simulación SIR')  # ← cada 5 elementos
    axs[0].set_ylabel('Fracción de población')
    axs[0].set_title(f'Control Óptimo Post-Confinamiento - {country}')
    axs[0].legend()
    axs[0].grid()

    # Gráfico inferior: Control H(t)
    axs[1].plot(dates[:-1], H, 'g--', label='Control H(t)')
    axs[1].set_xlabel('Fecha')
    axs[1].set_ylabel('Intensidad de control')
    axs[1].legend()
    axs[1].grid()

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# --- Datos de Italia ---
beta_italy = beta_opt #beta obtenido con el MODELO SIR antes de cuarentena
gamma_italy = gamma_opt #gamma obtenido con el MODELO SIR antes de cuarentena
_, I_italy, _, H_italy = simulate_sir_optimal_control(data_italy_pos, dates_italy_pos, beta_italy, gamma_italy)
plot_results(dates_italy_pos, data_italy_pos, I_italy, H_italy, "Italia")

I_italy_after = I_italy[-50:] #Datos del SIR luego de la cuarentena (utiles para luego comparar con los predichos)
```

# Creando una Feed Forward NN:

## Tomando los valores a partir de la cuarentena para predecir los siguientes 10 valores (es decir usando como entrenamiento los datos reales).

```python
def create_sequences(data, input_window, output_window):
    if len(data) < input_window + output_window:
        raise ValueError(f"Los datos son demasiado cortos ({len(data)}) para la ventana de entrada {input_window} y salida {output_window}")

    X, y = [], []
    for i in range(len(data) - input_window - output_window + 1):
        X.append(data[i:i+input_window])
        y.append(data[i+input_window:i+input_window+output_window])
    return np.array(X), np.array(y)

def build_fnn_model(input_window, output_window, hidden_units=64):
    model = Sequential([
        Input(shape=(input_window,)),
        Dense(hidden_units, activation='relu'),
        Dense(hidden_units, activation='relu'),
        Dense(hidden_units//2, activation='relu'),
        Dense(output_window)
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
    return model

def prepare_data(data, input_window, output_window, val_split_ratio=0.2):
    scaler = MinMaxScaler()
    data_normalized = scaler.fit_transform(data.reshape(-1, 1)).flatten()

    split_idx = int(len(data_normalized) * (1 - val_split_ratio))
    train_data = data_normalized[:split_idx]
    val_data = data_normalized[split_idx:]

    X_train, y_train = create_sequences(train_data, input_window, output_window)
    X_val, y_val = create_sequences(val_data, input_window, output_window)

    return X_train, y_train, X_val, y_val, scaler

def temporal_cross_validation(data, input_window=10, output_window=10, n_splits=3, epochs=100, batch_size=8):
    tscv = TimeSeriesSplit(n_splits=n_splits)
    scaler = MinMaxScaler()
    data_normalized = scaler.fit_transform(data.reshape(-1, 1)).flatten()

    fold_scores = []
    models = []

    for train_index, val_index in tscv.split(data_normalized):
        train_data = data_normalized[train_index]
        val_data = data_normalized[val_index]

        X_train, y_train = create_sequences(train_data, input_window, output_window)
        X_val, y_val = create_sequences(val_data, input_window, output_window)

        model = build_fnn_model(input_window, output_window)
        early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=[early_stop],
            verbose=0
        )

        val_loss = model.evaluate(X_val, y_val, verbose=0)[0]
        fold_scores.append(val_loss)
        models.append(model)

    best_model_idx = np.argmin(fold_scores)
    best_model = models[best_model_idx]

    return best_model, fold_scores, scaler

def train_and_predict(data, input_window=10, output_window=10, epochs=100, batch_size=8, val_split_ratio=0.2):
    X_train, y_train, X_val, y_val, scaler = prepare_data(
        data, input_window, output_window, val_split_ratio
    )

    model, fold_scores, _ = temporal_cross_validation(
        data, input_window, output_window, epochs=epochs, batch_size=batch_size
    )

    print(f"Pérdidas por fold: {fold_scores}")
    print(f"Pérdida promedio: {np.mean(fold_scores):.4f} ± {np.std(fold_scores):.4f}")

    last_seq = scaler.transform(data[-input_window:].reshape(-1, 1)).flatten()
    last_seq = last_seq.reshape(1, -1)
    prediction = model.predict(last_seq, verbose=0)
    prediction = scaler.inverse_transform(prediction.reshape(-1, 1)).flatten()

    return prediction, model, scaler

def plot_predictions(data, pred_10, true_future=None, input_window=10, titulo="Predicción"):
    """
    Args:
        data: array con los datos reales hasta el momento
        pred_10: predicción directa (10 días)
        true_future: array opcional con los datos reales futuros para comparar (10 días)
        input_window: tamaño de la ventana de entrada
        titulo: título del gráfico
    """
    # Últimos 10 días reales antes de la predicción
    real_base = data[-input_window:]

    # Ejes temporales
    days_real = np.arange(-len(real_base), 0)
    days_pred_10 = np.arange(0, len(pred_10))

    fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    # --- Gráfico superior: predicción vs realidad ---
    axs[0].plot(days_real, real_base, 'b-o', label="Datos reales anteriores", linewidth=2)
    axs[0].plot(days_pred_10, pred_10, 'g-x', label="Predicción 10 días", linewidth=2)

    if true_future is not None:
        axs[0].plot(days_pred_10, true_future, 'r--o', label="Datos reales futuros", linewidth=2)

    axs[0].axvline(0, color='k', linestyle=':', alpha=0.5)
    axs[0].set_title(titulo)
    axs[0].set_ylabel("Casos")
    axs[0].legend()
    axs[0].grid(True)

    # --- Gráfico inferior: error absoluto con línea punteada ---
    if true_future is not None:
        abs_error = np.abs(true_future - pred_10)
        axs[1].plot(days_pred_10, abs_error, 'g--o', label="Error absoluto", linewidth=2)
        axs[1].set_ylabel("Error absoluto")
        axs[1].set_xlabel("Días desde el último dato real")
        axs[1].set_title("Evolución del error absoluto")
        axs[1].legend()
        axs[1].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# 1. Italia
data = data_italy_pos.copy()
pred_10, model_italy, scaler = train_and_predict(data)
plot_predictions(data, pred_10, titulo="Predicción de casos - Italia", true_future=data_italy_after)

# 2. España
data = data_spain_pos.copy()
pred_10, model_spain, scaler = train_and_predict(data)
plot_predictions(data, pred_10, titulo="Predicción de casos - España", true_future=data_spain_after)

# 3. Colombia
data = data_colombia_pos.copy()
pred_10, model_colombia, scaler = train_and_predict(data)
plot_predictions(data, pred_10, titulo="Predicción de casos - Colombia", true_future=data_colombia_after)
```

## Ahora usando como entrenamiento los datos aumentados.

```python
## Configuración para datos aumentados
FACTOR_PER_DAY = 5  # 5 puntos por día
INPUT_WINDOW = 10 * FACTOR_PER_DAY  # 50 puntos (10 días)
OUTPUT_WINDOW = 10 * FACTOR_PER_DAY  # 50 puntos (10 días)

## Funciones específicas para datos aumentados
def prepare_synthetic_data(data, val_split_ratio=0.2):
    """Prepara datos con normalización y división temporal"""
    scaler = MinMaxScaler()
    data_norm = scaler.fit_transform(data.reshape(-1, 1)).flatten()

    split_idx = int(len(data_norm) * (1 - val_split_ratio))
    train_data = data_norm[:split_idx]
    val_data = data_norm[split_idx - INPUT_WINDOW - OUTPUT_WINDOW:]

    X_train, y_train = create_sequences(train_data, INPUT_WINDOW, OUTPUT_WINDOW)
    X_val, y_val = create_sequences(val_data, INPUT_WINDOW, OUTPUT_WINDOW)

    return X_train, y_train, X_val, y_val, scaler

def train_and_predict_synthetic(data, epochs=100, batch_size=16):
    """Entrena y predice para datos aumentados"""
    X_train, y_train, X_val, y_val, scaler = prepare_synthetic_data(data)

    model = build_fnn_model(INPUT_WINDOW, OUTPUT_WINDOW)
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop],
        verbose=0
    )

    # Solo predicción directa (no en cascada)
    last_seq = scaler.transform(data[-INPUT_WINDOW:].reshape(-1, 1)).flatten().reshape(1, -1)
    prediction = model.predict(last_seq, verbose=0)
    prediction = scaler.inverse_transform(prediction).flatten()

    return prediction, model, scaler

def plot_predictions_synthetic(data, pred_10_points, real_future=None, raw_future=None, titulo="Predicción"):
    """
    Grafica:
    - últimos datos aumentados
    - predicción del modelo
    - datos reales verdaderos (si se proporcionan)

    Args:
        data: datos aumentados originales
        pred_10_points: predicción del modelo (normalmente 50 puntos)
        real_future: datos aumentados futuros (sirve como comparación con predicción)
        raw_future: datos reales no aumentados para comparar con los aumentados
        titulo: título del gráfico
    """
    FACTOR_PER_DAY = 5
    PREDICTION_DAYS = len(pred_10_points) // FACTOR_PER_DAY
    REAL_DAYS = 5  # últimos días antes de predicción

    fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    # --- Gráfico 1: Datos y predicciones ---
    real_data = data[-REAL_DAYS * FACTOR_PER_DAY:]
    axs[0].plot(np.linspace(-REAL_DAYS, 0, len(real_data)), real_data, 'b-o',
                label="Datos reales (aumentados)", markersize=3, linewidth=1)

    axs[0].plot(np.linspace(0, PREDICTION_DAYS, len(pred_10_points)), pred_10_points, 'g-o',
                label="Predicción LSTM", linewidth=2)

    if real_future is not None:
        axs[0].plot(np.linspace(0, PREDICTION_DAYS, len(real_future)), real_future, 'r--x',
                    label="Datos futuros (aumentados)", linewidth=2)

    if raw_future is not None:
        axs[0].plot(np.linspace(0, PREDICTION_DAYS, len(raw_future)), raw_future, 'k-.s',
                    label="Datos reales (verdaderos)", linewidth=2)

    axs[0].axvline(0, color='k', linestyle=':', alpha=0.5)
    axs[0].set_title(titulo)
    axs[0].set_ylabel("Valor normalizado")
    axs[0].legend()
    axs[0].grid(True, alpha=0.3)

    # --- Gráfico 2: Error (opcional solo para real_future) ---
    if real_future is not None and len(real_future) == len(pred_10_points):
        error = np.abs(np.array(pred_10_points) - np.array(real_future))
        axs[1].plot(np.linspace(0, PREDICTION_DAYS, len(error)), error, 'g--', label="Error absoluto", linewidth=2)
        axs[1].set_ylabel("Error")
        axs[1].set_xlabel("Días desde el último dato real")
        axs[1].legend()
        axs[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
# ========= ITALIA =========
pred_10_italy, model_italy, scaler_italy = train_and_predict_synthetic(I_italy[:-50])
plot_predictions_synthetic(I_italy[:-50], pred_10_italy, real_future=I_italy_after, raw_future=data_italy_after, titulo="Italia - Datos aumentados")

# ========= ESPAÑA =========
pred_10_spain, model_spain, scaler_spain = train_and_predict_synthetic(I_spain[:-50])
plot_predictions_synthetic(I_spain[:-50], pred_10_spain, real_future=I_spain_after, raw_future=data_spain_after, titulo="España - Datos aumentados")

# ========= COLOMBIA =========
pred_10_colombia, model_colombia, scaler_colombia = train_and_predict_synthetic(I_colombia[:-50])
plot_predictions_synthetic(I_colombia[:-50], pred_10_colombia, real_future=I_colombia_after, raw_future=data_colombia_after, titulo="Colombia - Datos aumentados")
```

#Procedemos ahora a implementar una Non-linear autoregresive NN:

## Nuevamente, primero usamos los datos reales

```python
def build_nar_model(input_window, output_window, hidden_units=64):
    """Modelo NAR con profundidad similar a FNN (3 capas ocultas)"""
    model = Sequential([
        Input(shape=(input_window, 1)),  # Capa de entrada

        # Capa SimpleRNN (equivalente a primera capa oculta)
        SimpleRNN(hidden_units, activation='tanh', return_sequences=True),

        # Segunda capa SimpleRNN
        SimpleRNN(hidden_units//2, activation='tanh'),

        # Capas densas adicionales para igualar profundidad FNN
        Dense(hidden_units//2, activation='relu'),  # Tercera capa oculta
        Dense(hidden_units//4, activation='relu'),   # Cuarta capa oculta

        Dense(output_window)  # Capa de salida
    ])

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mae'])
    return model

def prepare_nar_data(data, input_window, output_window, val_split=0.2):
    """Prepara datos con normalización y división temporal adecuada"""
    scaler = MinMaxScaler()
    data_norm = scaler.fit_transform(data.reshape(-1, 1)).flatten()

    # División temporal sin superposición
    split_idx = int(len(data_norm) * (1 - val_split))
    train_data = data_norm[:split_idx]
    val_data = data_norm[split_idx:]  # Datos posteriores (sin overlap)

    # Crear secuencias manteniendo temporalidad
    X_train, y_train = create_sequences(train_data, input_window, output_window)
    X_val, y_val = create_sequences(val_data, input_window, output_window)

    # Reshape para RNN
    X_train = X_train.reshape(-1, input_window, 1)
    X_val = X_val.reshape(-1, input_window, 1)

    return X_train, y_train, X_val, y_val, scaler

def train_and_predict_nar(data, input_window=10, output_window=10, epochs=150):
    """Entrenamiento mejorado con validación temporal"""
    X_train, y_train, X_val, y_val, scaler = prepare_nar_data(
        data, input_window, output_window
    )

    model = build_nar_model(input_window, output_window)

    early_stop = EarlyStopping(monitor='val_loss',
                              patience=15,
                              restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=8,
        callbacks=[early_stop],
        verbose=0
    )

    # Predicción con normalización
    last_seq = scaler.transform(data[-input_window:].reshape(-1, 1))
    last_seq = last_seq.reshape(1, input_window, 1)
    prediction = model.predict(last_seq, verbose=0)
    prediction = scaler.inverse_transform(prediction).flatten()

    return prediction, model, scaler
```

```python
# Italia
pred_10_nar, model_italy_nar, scaler_italy = train_and_predict_nar(data_italy_pos)
plot_predictions(data_italy_pos, pred_10_nar, true_future=data_italy_after, titulo="Italia (NAR)")

# España
pred_10_nar, model_spain_nar, scaler_spain = train_and_predict_nar(data_spain_pos)
plot_predictions(data_spain_pos, pred_10_nar, true_future=data_spain_after, titulo="España (NAR)")

# Colombia
pred_10_nar, model_colombia_nar, scaler_colombia = train_and_predict_nar(data_colombia_pos)
plot_predictions(data_colombia_pos, pred_10_nar, true_future=data_colombia_after, titulo="Colombia (NAR)")
```

##Ahora con los datos aumentados

```python
def build_nar_augmented_model(input_window, output_window, hidden_units=64):
    """Modelo NAR profundo para datos aumentados"""
    model = Sequential([
        Input(shape=(input_window, 1)),

        # Capas recurrentes
        SimpleRNN(hidden_units, activation='tanh', return_sequences=True),
        SimpleRNN(hidden_units//2, activation='tanh'),

        # Capas densas (igualando profundidad FNN)
        Dense(hidden_units//2, activation='relu'),
        Dense(hidden_units//4, activation='relu'),

        Dense(output_window)
    ])
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mae'])
    return model

def prepare_nar_augmented_data(data, val_split=0.2):
    scaler = MinMaxScaler()
    data_norm = scaler.fit_transform(data.reshape(-1, 1)).flatten()

    split_idx = int(len(data_norm) * (1 - val_split))
    train_data = data_norm[:split_idx]
    val_data = data_norm[split_idx:]

    # Creación de secuencias con reshape integrado
    X_train = np.array([train_data[i:i+INPUT_WINDOW] for i in range(len(train_data)-INPUT_WINDOW-OUTPUT_WINDOW+1)])
    X_val = np.array([val_data[i:i+INPUT_WINDOW] for i in range(len(val_data)-INPUT_WINDOW-OUTPUT_WINDOW+1)])

    # Reshape directo a 3D (más eficiente)
    X_train = X_train.reshape(-1, INPUT_WINDOW, 1)
    X_val = X_val.reshape(-1, INPUT_WINDOW, 1)

    # Targets (sin reshape necesario)
    y_train = np.array([train_data[i+INPUT_WINDOW:i+INPUT_WINDOW+OUTPUT_WINDOW] for i in range(len(train_data)-INPUT_WINDOW-OUTPUT_WINDOW+1)])
    y_val = np.array([val_data[i+INPUT_WINDOW:i+INPUT_WINDOW+OUTPUT_WINDOW] for i in range(len(val_data)-INPUT_WINDOW-OUTPUT_WINDOW+1)])

    return X_train, y_train, X_val, y_val, scaler

def train_and_predict_nar_augmented(data, epochs=150):
    """Entrenamiento para datos aumentados"""
    X_train, y_train, X_val, y_val, scaler = prepare_nar_augmented_data(data)

    model = build_nar_augmented_model(INPUT_WINDOW, OUTPUT_WINDOW)

    early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=16,
        callbacks=[early_stop],
        verbose=0
    )

    # Predicción
    last_seq = scaler.transform(data[-INPUT_WINDOW:].reshape(-1, 1))
    last_seq = last_seq.reshape(1, INPUT_WINDOW, 1)
    prediction = model.predict(last_seq, verbose=0)
    prediction = scaler.inverse_transform(prediction).flatten()

    return prediction, model, scaler
```

```python
# Italia
pred_10_italy, model_italy, scaler_italy = train_and_predict_nar_augmented(I_italy[:-50])
plot_predictions_synthetic(I_italy[:-50], pred_10_italy, real_future=I_italy_after, raw_future=data_italy_after, titulo="Italia - NAR (datos aumentados)")

# España
pred_10_spain, model_spain, scaler_spain = train_and_predict_nar_augmented(I_spain[:-50])
plot_predictions_synthetic(I_spain[:-50], pred_10_spain, real_future=I_spain_after, raw_future=data_spain_after, titulo="España - NAR (datos aumentados)")

# Colombia
pred_10_colombia, model_colombia, scaler_colombia = train_and_predict_nar_augmented(I_colombia[:-50])
plot_predictions_synthetic(I_colombia[:-50], pred_10_colombia, real_future=I_colombia_after, raw_future=data_colombia_after, titulo="Colombia - NAR (datos aumentados)")
```

# Ahora implementamos una LSTM, mucho más util a la hora de predecir lineas de tiempo:

##Con los datos reales

```python
def build_lstm_model(input_window, output_window):
    model = Sequential([
        Input(shape=(input_window, 1)),
        LSTM(64, activation='tanh'),
        Dense(32, activation='relu'),
        Dense(output_window)
    ])
    model.compile(optimizer=Adam(0.001), loss='mse', metrics=['mae'])
    return model

def prepare_lstm_data(data, input_window, output_window, val_split=0.2):
    scaler = MinMaxScaler()
    data_norm = scaler.fit_transform(data.reshape(-1, 1)).flatten()
    split_idx = int(len(data_norm) * (1 - val_split))

    train_data = data_norm[:split_idx]
    val_data = data_norm[split_idx - input_window:]

    X_train, y_train = create_sequences(train_data, input_window, output_window)
    X_val, y_val = create_sequences(val_data, input_window, output_window)

    return X_train, y_train, X_val, y_val, scaler

def train_and_predict_lstm(data, input_window=10, output_window=10, epochs=150):
    X_train, y_train, X_val, y_val, scaler = prepare_lstm_data(data, input_window, output_window)

    model = build_lstm_model(input_window, output_window)
    early_stop = EarlyStopping(patience=10, restore_best_weights=True)

    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=8,
        callbacks=[early_stop],
        verbose=0
    )

    # Predicción directa
    last_seq = scaler.transform(data[-input_window:].reshape(-1, 1)).reshape(1, input_window, 1)
    pred = scaler.inverse_transform(model.predict(last_seq, verbose=0)).flatten()
    return pred, model, scaler
```

```python
# Italia
pred_10_ita, model_ita, scaler_ita = train_and_predict(data_italy_pos)
plot_predictions(data_italy_pos, pred_10_ita, true_future=data_italy_after, titulo="Italia (LSTM)")

# España
pred_10_esp, model_esp, scaler_esp = train_and_predict(data_spain_pos)
plot_predictions(data_spain_pos, pred_10_esp, true_future=data_spain_after, titulo="España (LSTM)")

# Colombia
pred_10_col, model_col, scaler_col = train_and_predict(data_colombia_pos)
plot_predictions(data_colombia_pos, pred_10_col, true_future=data_colombia_after, titulo="Colombia (LSTM)")
```

## Y con los aumentados

```python
# Configuración aumentada
FACTOR_PER_DAY = 5
INPUT_WINDOW = 10 * FACTOR_PER_DAY  # 50 puntos
OUTPUT_WINDOW = 10 * FACTOR_PER_DAY  # 50 puntos

def create_sequences_augmented(data, input_window, output_window):
    """Crea secuencias para datos aumentados con reshape 3D"""
    X, y = [], []
    for i in range(len(data) - input_window - output_window + 1):
        X.append(data[i:i+input_window])
        y.append(data[i+input_window:i+input_window+output_window])
    return np.array(X).reshape(-1, input_window, 1), np.array(y)  # (muestras, timesteps, features)

def build_lstm_augmented_model(input_window, output_window):
    """Modelo LSTM para datos de alta frecuencia"""
    model = Sequential([
        Input(shape=(input_window, 1)),
        LSTM(128, activation='tanh', return_sequences=True),
        LSTM(64, activation='tanh'),
        Dense(64, activation='relu'),
        Dense(output_window)
    ])
    model.compile(optimizer=Adam(0.0005), loss='mse', metrics=['mae'])
    return model

def train_and_predict_augmented(data, input_window=50, output_window=50, epochs=200):
    """Pipeline completo para datos aumentados"""
    # Normalización
    scaler = MinMaxScaler()
    data_norm = scaler.fit_transform(data.reshape(-1, 1)).flatten()

    # División train/val (últimos 20% para validación)
    split_idx = int(len(data_norm) * 0.8)
    X_train, y_train = create_sequences_augmented(data_norm[:split_idx], input_window, output_window)
    X_val, y_val = create_sequences_augmented(data_norm[split_idx-input_window:], input_window, output_window)

    # Entrenamiento
    model = build_lstm_augmented_model(input_window, output_window)
    model.fit(X_train, y_train,
              validation_data=(X_val, y_val),
              epochs=epochs,
              batch_size=16,
              callbacks=[EarlyStopping(patience=20, restore_best_weights=True)],
              verbose=0)

    # Predicción
    last_seq = scaler.transform(data[-input_window:].reshape(-1, 1)).reshape(1, input_window, 1)
    pred = scaler.inverse_transform(model.predict(last_seq, verbose=0)).flatten()
    return pred, model, scaler
```

```python
# Italia
pred_10_ita, model_ita, scaler_ita = train_and_predict_augmented(I_italy[:-50], INPUT_WINDOW, OUTPUT_WINDOW)
plot_predictions_synthetic(I_italy[:-50], pred_10_ita, real_future=I_italy_after, raw_future=data_italy_after, titulo="Italia (LSTM aumentado)")

# España
pred_10_esp, model_esp, scaler_esp = train_and_predict_augmented(I_spain[:-50], INPUT_WINDOW, OUTPUT_WINDOW)
plot_predictions_synthetic(I_spain[:-50], pred_10_esp, real_future=I_spain_after, raw_future=data_spain_after, titulo="España (LSTM aumentado)")

# Colombia
pred_10_col, model_col, scaler_col = train_and_predict_augmented(I_colombia[:-50], INPUT_WINDOW, OUTPUT_WINDOW)
plot_predictions_synthetic(I_colombia[:-50], pred_10_col, real_future=I_colombia_after, raw_future=data_colombia_after, titulo="Colombia (LSTM aumentado)")
```
